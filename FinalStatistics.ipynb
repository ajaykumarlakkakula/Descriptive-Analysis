```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv("Downloads/cybersecurity_attacks.csv")
```

```python
##Loading the data set using python library pandas
df=pd.DataFrame(data)
df
```

```python
##Displays the top five rows
df.head()
```

```python
##It will help to understand number of rows and columns, null values, data types and complete structure of data 
df.info()
```

```python
##Gives information of rows and columns
df.shape
```

```python
df.columns 
```

```python
##Tells us which column contains null values
df.isnull().sum().sort_values(ascending=False)
```

```python
## It helps you understand the different types of alerts or warnings in your dataset.
df['Alerts/Warnings'].unique()
```

```python
##filling Nan columns with None
df['Alerts/Warnings'] = df['Alerts/Warnings'].fillna('None')
df['Alerts/Warnings']
```

```python
##filling Nan columns with No Data
df['IDS/IPS Alerts'] = df['IDS/IPS Alerts'].fillna('No Data')
df['IDS/IPS Alerts']
```

```python
##filling Nan columns with No Detection
df['Malware Indicators'] = df['Malware Indicators'].fillna('No Detection')
df['Malware Indicators']
```

```python
##filling Nan columns with No Data
df['Firewall Logs'] = df['Firewall Logs'].fillna('No Data')
df['Firewall Logs']
```

```python
##filling Nan columns with '0'
df['Proxy Information'] = df['Proxy Information'].fillna(0)
df['Proxy Information']
```

```python
#provides a count of missing values in each column of the DataFrame.
df.isnull().sum()
```

```python
# Descriptive statistics for numerical columns
numerical_summary = df.describe()
print(numerical_summary)
```

```python
# Frequency distribution for categorical columns
categorical_summary = df['Attack Type'].value_counts().sort_index()
print(categorical_summary)
```

```python
## Convert Timestamp to Datetime Format:
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df['Timestamp']
```

```python
##Extracting and creating columns Year,Month,Day,Hour,Minute,Second from Timestamp
df['Year'] = df['Timestamp'].dt.year
df['Month'] = df['Timestamp'].dt.month
df['Day'] = df['Timestamp'].dt.day
df['Hour'] =df['Timestamp'].dt.hour
df['Minute']= df['Timestamp'].dt.minute
df['Second'] = df['Timestamp'].dt.second
df
```

```python
##Below command gives information of Attack types year wise by using group by.
df.groupby(['Year'])['Attack Type'].value_counts()
```

```python
##Below command gives details of column names in the dataset
df.columns
```

```python
###Distribution of Key Categorical Variables in Cybersecurity Attacks
##These count plots help you understand the distribution of categories, their frequencies, and overall patterns in the dataset.
plt.figure(figsize = (12,12))

##subplot 1
plt.subplot(3,2,1)
sns.countplot(data=df, x='Attack Type', order=df['Attack Type'].value_counts().index, palette='viridis')
plt.title('Attack Types distribution')
plt.xlabel('Attack Type')
plt.ylabel('Frequency')
plt.xticks(rotation=45)

##subplot 2
###Severity Levels distribution
plt.subplot(3,2,2)
sns.countplot(data=df, x='Severity Level', order=df['Severity Level'].value_counts().index, palette='plasma')
plt.title('Severity Levels distribution')
plt.xlabel('Severity Level')
plt.ylabel('Frequency')
plt.xticks(rotation=45)

##subplot 3
##Protocol distribution
plt.subplot(3,2,3)
sns.countplot(data=df, x= 'Protocol',order = df['Protocol'].value_counts().index, palette ='pastel')
plt.title('Protocol distribution')
plt.xlabel('Protocol')
plt.ylabel('Frequency')
plt.xticks(rotation=45)

##subplot 4
###Action Taken distribution
plt.subplot(3,2,4)
sns.countplot(data=df, x= 'Action Taken', order = df['Action Taken'].value_counts().index, palette ='Set1')
plt.title('Action Taken distribution')
plt.xlabel('Action Taken')
plt.ylabel('Frequency')
plt.xticks(rotation=45)

##subplot 5
plt.subplot(3,2,5)
sns.countplot(data=df, x= 'Traffic Type', order = df['Traffic Type'].value_counts().index, palette ='Set1')
plt.title('Traffic Type distribution')
plt.xlabel('Traffic Type')
plt.ylabel('Frequency')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
##Distribution Analysis of Key Numerical Features in Cybersecurity Data


numerical_columns =['Source Port','Packet Length','Anomaly Scores']

for i in numerical_columns:
    plt.figure(figsize=(12,6))
    n = len(df[i])  # Number of data points
    num_bins_sqrt = int(np.ceil(np.sqrt(n)))
    sns.histplot(df[i],bins = num_bins_sqrt, color= 'brown', edgecolor='black',alpha=0.7,kde=True)
    mean_value = df[i].mean()
    median_value = df[i].median()
    plt.axvline(mean_value, color='red', linestyle='--', label='Mean')
    plt.axvline(median_value, color='blue', linestyle='-', label='Median')
    plt.title(f'Distribution of {i}')
    plt.xlabel(i)
    plt.ylabel('Frequency')
    plt.legend(loc='upper left', bbox_to_anchor=(1, 1), frameon=False)
    plt.tight_layout()
    plt.show()
    
```

```python
##  From  above it's understandable that the mean line may be difficult to distinguish in the histogram for packet 
mean_value = df['Packet Length'].mean()
median_value = df['Packet Length'].median()
print(f"Mean Packet Length: {mean_value}")
print(f"Median Packet Length: {median_value}")
```

```python
##Geo-location of Attacks(Top 10)
plt.figure(figsize=(10,6))
top_locations = df['Geo-location Data'].value_counts().head(10)
sns.barplot(x=top_locations.index, y=top_locations.values, palette='mako')
plt.title('Top 10 Locations of Attacks')
plt.xlabel('Location')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
##Attack over years
plt.figure(figsize=(12,6))
sns.countplot(data=df, x='Year', hue='Attack Type', palette='bright')
plt.title('Attack Types Distribution Over Years')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.legend(title='Attack Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
##Traffic type distribution 
plt.figure(figsize=(8,6))
sns.countplot(data=df, x= 'Year', hue ='Traffic Type', palette ='dark')
plt.title('Traffic type distribution over years')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Protocol', hue='Severity Level', palette='mako')
plt.title('Protocol Distribution by Severity Level')
plt.xlabel('Protocol')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.legend(title='Severity Level',bbox_to_anchor =(1,1))
plt.show()
```

```python
#A violin plot can show the distribution of the data along with its probability density.
plt.figure(figsize=(12, 6))
sns.violinplot(data=df, x='Action Taken', y='Anomaly Scores', palette='muted')
plt.title('Anomaly Scores Distribution by Action Taken')
plt.xlabel('Action Taken')
plt.ylabel('Anomaly Scores')
plt.xticks(rotation=45)
plt.show()
```
